```python
data=pd.read_csv('drive/MyDrive/Movies1.csv')
```

```python
data.drop(['index'], axis=1)
```

```python
data.shape
```

```python
data.drop(['id', 'index'], axis=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
relevant_features= ['genres', 'production_companies', 'director']
```

```python
for features in relevant_features:
  data[features]=data[features].fillna('')
```

```python
combined_features = data['genres'] + ' ' + data['production_companies'] + ' ' + data['director']
```

```python
print(combined_features)
```

```python
vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(feature_vectors)
print(similarity)
```

```python
print(similarity.shape)
```

```python
movie_name=str(input('Enter the movie name: '))
```

```python
list_of_all_movies=str(data['original_title'])
print(list_of_all_movies)
```

```python
import numpy as np
import difflib
```

```python
numpy_array=np.array(list(list_of_all_movies))
```

```python
closest_match = difflib.get_close_matches(movie_name, list_of_all_movies, cutoff=0.0000000007)
print(closest_match)
```

```python
close_match=closest_match[0]
print(close_match)
```

```python
index_of_the_movie = data[data.original_title == close_match]['index'].values
print(index_of_the_movie)
```

```python
index_of_the_movie=0
similarity_score=list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sort_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sort_similar_movies)
```

```python
print('Movies suggested for you: \n')
i = 1

for movies in sort_similar_movies:
    index = movies[0]
    title_from_index = data[data.index==index]['original_title'].values[0]
    if(i<30):
        print(i,'.',title_from_index)
        i+=1
```
