```python
import matplotlib.pyplot as plt
five_star_products=data[data['Rate']==5]
products=five_star_products['product_name'].value_counts()
plt.pie(products, labels=products.index, autopct='%1.1f%%')
plt.show()
```

```python
five_star_products=data[data['Rate']==1]
products=five_star_products['product_name'].value_counts()
plt.pie(products, labels=products.index, autopct='%1.1f%%')
plt.show()
```

Based on the above charts, we can draw followng insights:

1. Black Women Sling Bag is the best product from the list pf products, as it has maximum number of 5-star Ratings and least number of 1-star Ratings.
2. Women Black Sling bag and Women White Sling bags are the products that have more 1-star Ratings than 5-star Ratings. Hence, those products aren't recommended.
